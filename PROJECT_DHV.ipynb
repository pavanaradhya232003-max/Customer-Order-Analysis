```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (8, 5)

# LOADING THE DATASETS
orders = pd.read_csv("orders_dataset_updated.xls")

orders.head()
```

```python
customers = pd.read_csv("customers_dataset_updated.xls")
customers.head()
```

```python
print(f"Orders shape: {orders.shape}")
print(f"Customers shape: {customers.shape}")
```

```python
# FINDING THE NULL VALUES

print("Orders null values:")
print(orders.isnull().sum())
print(f"\nTotal nulls in orders: {orders.isnull().sum().sum()}")

print("\nCustomers null values:")
print(customers.isnull().sum())
print(f"\nTotal nulls in customers: {customers.isnull().sum().sum()}")
```

```python
# DUPLICATE ROWS
print(f"Duplicate rows in orders: {orders.duplicated().sum()}")
print(f"Duplicate rows in customers: {customers.duplicated().sum()}")
```

```python
# DATA CLEANING & DROPING THE DUPLICATES

orders_clean = orders.copy()
orders_clean["Review_Rating"] = orders_clean["Review_Rating"].fillna(orders_clean["Review_Rating"].median())
orders_clean["Price"] = orders_clean["Price"].fillna(orders_clean["Price"].median())
orders_clean["Quantity"] = orders_clean["Quantity"].fillna(orders_clean["Quantity"].median())
orders_clean["Delivery_Status"] = orders_clean["Delivery_Status"].fillna(orders_clean["Delivery_Status"].mode()[0])
orders_clean["Product_Category"] = orders_clean["Product_Category"].fillna("Unknown")
orders_clean["Review_Text"] = orders_clean["Review_Text"].fillna("No Review")

orders_clean = orders_clean.drop_duplicates()

print(f"Orders shape before: {orders.shape} -> after: {orders_clean.shape}")
print(f"Remaining nulls: {orders_clean.isnull().sum().sum()}")
print(f"Remaining duplicates: {orders_clean.duplicated().sum()}")
```

```python
customers_clean = customers.copy()
customers_clean["Customer_ID"] = customers_clean["Customer_ID"].fillna("Unknown_ID")
customers_clean["Customer_Name"] = customers_clean["Customer_Name"].fillna("Unknown")
customers_clean["Payment_Method"] = customers_clean["Payment_Method"].fillna(customers_clean["Payment_Method"].mode()[0])
customers_clean["State"] = customers_clean["State"].fillna("Unknown")

customers_clean = customers_clean.drop_duplicates()

print(f"Customers shape before: {customers.shape} -> after: {customers_clean.shape}")
print(f"Remaining nulls: {customers_clean.isnull().sum().sum()}")
print(f"Remaining duplicates: {customers_clean.duplicated().sum()}")
```

```python
#SAVING CLEANED DATASET
orders_clean.to_csv("orders.csv", index=False)
customers_clean.to_csv("customers.csv", index=False)
```
